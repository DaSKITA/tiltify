```python
from tiltify.data_structures.document_collection import DocumentCollection
from tiltify.parsers.policy_parser import PolicyParser
from tiltify.data_loader import DataLoader
from tiltify.preprocessing.label_retriever import LabelRetriever

webde_path = "/home/gebauer/Desktop/repo/tiltify/data/annotated_policies/webde.json"
data_loader = DataLoader()
policies = data_loader.get_json_policies("annotated_policies")

doc = [policy for policy in policies if policy["document"]["document_name"] == "webde"][0]
doc["annotations"]
```

```python
parser = PolicyParser()
label_retriever = LabelRetriever()
parsed_doc = parser.parse(doc["document"]["document_name"], doc["document"]["text"], doc["annotations"])
annotations = label_retriever.get_annotations(parsed_doc.blobs)
print([annotation[0].label for annotation in annotations if annotation != []])
```

```python
labels = label_retriever.retrieve_labels(parsed_doc.blobs)
print([label for label in labels if label[0] > 0 ])
```
