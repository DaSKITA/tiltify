# Create Latex Table

```python
import os
from tiltify.config import Path
from glob import glob
import json

iwpe_dir = os.path.join(Path.root_path, "experiments/IWPE/results_2")
results_per_model = {}
for results_path in glob(os.path.join(iwpe_dir, "*/right_to_results.json")):
    model_name = results_path.split("/")[-2]
    with open(results_path, "r") as f:
        results = json.load(f)
    results_per_model[model_name] = results
```

```python
import pandas as pd

def shape_results(res_dict):
    metrics = []
    for type_of_res, type_res in res_dict.items():
        if type_of_res != "classify_metrics":
            classify_matrix = pd.DataFrame.from_dict([type_res["True"]])
        else:
            classify_matrix = pd.DataFrame.from_dict([type_res["macro avg"]])
        classify_matrix.loc[:, "res_type"] = type_of_res
        classify_matrix.columns = [f"{type_of_res}_{col_name}" for col_name in classify_matrix.columns]
        metrics.append(classify_matrix)
    return pd.concat(metrics)

def mean_std(df):
    std_df = df.std()
    mean_df = df.mean()
    for col_idx, cell_mean in mean_df.iteritems():
        std_df.loc[col_idx] = f"{round(cell_mean, 2)} ({round(std_df.loc[col_idx], 2)})"
    return std_df
    

def create_right_to_df(results):
    train_right_df = []
    test_right_df = []
    for model_name, model_res in results.items():
        for right_to, right_to_res in model_res.items():
            training = []
            test = []
            for k, k_entry in right_to_res.items():
                train_res = shape_results(k_entry["train_results"])
                test_res = shape_results(k_entry["test_results"])
                
                training.append(train_res)
                test.append(test_res)
            all_test = mean_std(pd.concat(test))
            all_train = mean_std(pd.concat(training))
            all_train["model"] = model_name
            all_train["right_to"] = right_to
            all_test["model"] = model_name
            all_test["right_to"] = right_to
            train_right_df.append(all_train)
            test_right_df.append(all_test)
    return pd.concat(train_right_df, axis=1).transpose().set_index(keys=["right_to"]), pd.concat(test_right_df, axis=1).transpose().set_index(keys=["right_to"])

```

```python
train_right_to, test_right_to = create_right_to_df(results_per_model)
test_right_to
```
